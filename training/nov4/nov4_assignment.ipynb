```python
# Numbers (Operations and Methods) 

# 1. Integer Operations 
# ○ Write a Python program to add two integers. Input: a = 5, b = 3. Output: 
# Sum: 8.
a = 5
b = 3
sum = a+b
print("Sum:", sum)
```

```python
# 2. Float Operations 
# ○ Write a program to multiply two float numbers. Input: x = 2.5, y = 1.5. 
# Output: Product: 3.75.
x = 2.5
y = 1.5
print("Product: ", x*y)
```

```python
# 3. Complex Number Operations 
# ○ Create a Python function to add two complex numbers. Input: z1 = 2+3j, z2 
# = 1-1j. Output: Sum: (3+2j).
def add(z1, z2):
    result = z1 + z2
    return result
    
z1 = 2+3j
z2 = 1-1j

sum = add(z1,z2)

print("Sum: ", sum)
```

```python
# 4. Integer Division 
# ○ Write a program to perform integer division of two numbers. Input: a = 7, b = 
# 2. Output: Quotient: 3.
a = 7
b = 2
div = a//b
print("Quotient: ", div)
```

```python
# 5. Power Calculation 
# ○ Create a function to calculate the power of a number. Input: base = 2, 
# exponent = 3. Output: Result: 8.
def cal_power(base, exp):
    res = base ** exp
    return res
base = 2
exp = 4

result = cal_power(base, exp)

print("Result: " , result)
```

```python
# 6. Float Formatting 
# ○ Write a program to display a float number with 2 decimal places. Input: num = 
# 3.14159. Output: Formatted number: 3.14.
num = 22/7
print(f"Formatted number: {num:.2f}")
```

```python
# 7. Complex Conjugate 
# ○ Write a Python function to find the conjugate of a given complex number. Input: z 
# = 4-3j. Output: Conjugate: (4+3j).
z = 4-3j
print("Conjugate: ", z.conjugate())
```

```python
# 8. Absolute Value 
# ○ Create a program to find the absolute value of an integer. Input: -5. Output: 
# Absolute value: 5.
num = -5
abs_value = -(num)
print("Absolute value:", abs_value)
```

```python
# 9. Float Comparison 
# ○ Write a program to compare two float numbers and check if they are 
# approximately equal (considering a tolerance of 0.001). Input: x = 1.234, y = 
# 1.235. Output: Approximately equal.
x = 1.234
y = 1.235

tolerance = 0.001

if abs(x - y) <= tolerance:
    print("Approximately equal")
else:
    print("Not equal")
```
